```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Specify the year for debugging
year = 1983

# URL for the race results page of the year
url = f'https://www.formula1.com/en/results.html/{year}/races.html'

# Send a GET request to fetch the HTML content
headers = {'User-Agent': 'Mozilla/5.0'}  # Mimic a browser
r = requests.get(url, headers=headers)

# Parse the HTML content
soup = BeautifulSoup(r.text, 'html.parser')

# Find all race links for the year
year_links = []
for page in soup.find_all('a', attrs={'class': "block"}):  # Look for <a> tags with class "block"
    link = page.get('href')
    if f'/en/results/{year}/races/' in link:  # Ensure it matches the race results pattern
        year_links.append(link)

# Debug: Print the extracted links
print(f"Race Links for {year}: {year_links}")

# Initialize a DataFrame for the year
year_df = pd.DataFrame()

# Base URL for qualifying results
base_url = 'https://www.formula1.com'

# Iterate through race links
for n, link in enumerate(year_links):
    # Construct the qualifying results URL
    qualifying_link = link.replace('race-result', 'starting-grid')  # Replace 'race-result' with 'starting-grid'
    full_url = f"{base_url}{qualifying_link}"

    print(f"Fetching data from: {full_url}")  # Debug: Print the full URL

    try:
        # Attempt to read the qualifying table from the page
        dfs = pd.read_html(full_url)  # Returns a list of tables
        df = dfs[0]  # First table is usually the qualifying results

        # Add season and round information
        df['season'] = year
        df['round'] = n + 1

        # Remove unnecessary columns
        for col in df.columns:
            if 'Unnamed' in col:
                df.drop(col, axis=1, inplace=True)

        # Append to the year DataFrame
        year_df = pd.concat([year_df, df], ignore_index=True)

    except Exception as e:
        # Handle exceptions (e.g., no table found)
        print(f"Error fetching data for race {n+1}: {e}")

# Output the shape and preview the data for the year
print(f"Year {year} DataFrame Shape: {year_df.shape}")
print(year_df.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the range of years
start_year = 1980
end_year = 2022

# Initialize a DataFrame to store all years' data
all_qualifying_results = pd.DataFrame()

# Base URL for qualifying results
base_url = 'https://www.formula1.com'

# Loop through each year
for year in range(start_year, end_year + 1):
    print(f"Processing year: {year}")

    # URL for the race results page of the year
    url = f'https://www.formula1.com/en/results.html/{year}/races.html'

    # Send a GET request to fetch the HTML content
    headers = {'User-Agent': 'Mozilla/5.0'}  # Mimic a browser
    r = requests.get(url, headers=headers)

    # Parse the HTML content
    soup = BeautifulSoup(r.text, 'html.parser')

    # Find all race links for the year
    year_links = []
    for page in soup.find_all('a', attrs={'class': "block"}):  # Look for <a> tags with class "block"
        link = page.get('href')
        if f'/en/results/{year}/races/' in link:  # Ensure it matches the race results pattern
            year_links.append(link)

    # Debug: Print the extracted links
    print(f"Race Links for {year}: {year_links}")

    # Initialize a DataFrame for the current year
    year_df = pd.DataFrame()

    # Iterate through race links
    for n, link in enumerate(year_links):
        # Construct the qualifying results URL
        qualifying_link = link.replace('race-result', 'starting-grid')  # Replace 'race-result' with 'starting-grid'
        full_url = f"{base_url}{qualifying_link}"

        print(f"Fetching data from: {full_url}")  # Debug: Print the full URL

        try:
            # Attempt to read the qualifying table from the page
            dfs = pd.read_html(full_url)  # Returns a list of tables
            df = dfs[0]  # First table is usually the qualifying results

            # Add season and round information
            df['season'] = year
            df['round'] = n + 1

            # Remove unnecessary columns
            for col in df.columns:
                if 'Unnamed' in col:
                    df.drop(col, axis=1, inplace=True)

            # Append to the year DataFrame
            year_df = pd.concat([year_df, df], ignore_index=True)

        except Exception as e:
            # Handle exceptions (e.g., no table found)
            print(f"Error fetching data for race {n+1} in {year}: {e}")

    # Append the current year's data to the consolidated DataFrame
    all_qualifying_results = pd.concat([all_qualifying_results, year_df], ignore_index=True)

    # Debug: Print the year's DataFrame shape
    print(f"Year {year} DataFrame Shape: {year_df.shape}")

# Output the shape and save the results to a CSV file
print(f"Total DataFrame Shape: {all_qualifying_results.shape}")

```
